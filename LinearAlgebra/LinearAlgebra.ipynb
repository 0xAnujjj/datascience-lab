```python
# import
import numpy as np
import numpy.linalg as LA
```

### 1. Matrix Operation

```python
A = np.array([[2, -1], [0, 3]])
B= np.array([[1, 2], [-3, 4]])

# A + B
add_result = A + B

# AB
mul_result = A @ B

# 3A - 2B
exp_result = 3*A - 2*B

#print the results
print("A + B:\n", add_result)
print("\nAB:\n", mul_result)
print("\n3A - 2B:\n", exp_result)
```

### 2. Determinants and Inverses

```python
# Determinant of A
det_A = LA.det(A)
print("Determinant of A:", det_A)

if det_A != 0:
    inv_A = LA.inv(A)
    print("Matrix A is invertible and")
    print("\nInverse of A:\n", inv_A)
else:
    print("\nMatrix A is not invertible")
```

**Since the determinant of A isn't equal to zero (det(A) = 6), the matrix A is invertible.**

### 4. Car Price Prediction

```python
#x1 x2 x3 price
x = np.array([[4, 1.5, 100],
                [6, 2.0, 150]])

#price vector
y = np.array([20, 30])

x_transpose = x.T
beta = LA.pinv(x_transpose @ x) @ x_transpose @ y

print('\n Coefficient vector(beta):\n', beta)

#predict price for third car
x_new = np.array([8, 3.0, 200])
predicted_price = x_new @ beta

print("\nPredicted price of car 3:", predicted_price)
```

**So, the predicted price of car 3 is: 39.99**
